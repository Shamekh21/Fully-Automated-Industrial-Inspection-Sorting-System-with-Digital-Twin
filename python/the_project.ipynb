## 📂 تحديد مسارات البيانات
### مجموعة بيانات Casting Defect Detection
تحتوي مجموعة البيانات على صور لـ**أجزاء مصبوبة** (Casting Parts):
- **ok_front**: قطع سليمة بدون عيوب
- **def_front**: قطع معيبة (خدوش، شقوق، تشوهات)

**حجم الصور**: 300×300 بكسل تقريبًا  
**نوع الملفات**: JPEG / PNG


````python
# ============================================================
# تعريف مسارات مجلدات التدريب
# ============================================================
# مجلد الصور السليمة (OK Parts)
folder_ok = r"E:\AI\AI project\code\archive\temp\casting_data\casting_data\train\ok_front"

# مجلد الصور المعيبة (Defective Parts)
folder_defect = r"E:\AI\AI project\code\archive\temp\casting_data\casting_data\train\def_front"

print(f"📁 مجلد القطع السليمة : {folder_ok}")
print(f"📁 مجلد القطع المعيبة : {folder_defect}")
````

````python
# ============================================================
# التحقق من وجود المجلدات وتحميل الصور
# ============================================================
try:
    # التحقق من وجود المجلدين
    if not os.path.isdir(folder_ok):
        raise FileNotFoundError(f"مجلد القطع السليمة غير موجود: {folder_ok}")
    if not os.path.isdir(folder_defect):
        raise FileNotFoundError(f"مجلد القطع المعيبة غير موجود: {folder_defect}")

    # قراءة أسماء ملفات الصور من المجلدين (فلترة الامتدادات المدعومة فقط)
    extensions = ('.png', '.jpg', '.jpeg', '.bmp')
    ok_files     = sorted([f for f in os.listdir(folder_ok)     if f.lower().endswith(extensions)])
    defect_files = sorted([f for f in os.listdir(folder_defect) if f.lower().endswith(extensions)])

    print(f"✅ عدد صور القطع السليمة : {len(ok_files)}")
    print(f"✅ عدد صور القطع المعيبة : {len(defect_files)}")

    # التحقق من وجود صور فعلية داخل المجلدات
    if not ok_files:
        raise FileNotFoundError(f"لا توجد صور في المجلد: {folder_ok}")
    if not defect_files:
        raise FileNotFoundError(f"لا توجد صور في المجلد: {folder_defect}")

    # بناء المسار الكامل لأول صورة من كل مجلد
    path_ok     = os.path.join(folder_ok,     ok_files[0])
    path_defect = os.path.join(folder_defect, defect_files[0])

    print(f"\n📷 سيتم تحميل الصورة السليمة    : {ok_files[0]}")
    print(f"📷 سيتم تحميل الصورة المعيبة    : {defect_files[0]}")

except FileNotFoundError as e:
    print(f"❌ خطأ: {e}")
    print("💡 تلميح: تأكد من تشغيل سكريبت oragnize.py أولاً لتنظيم الملفات")
````

## 🖼️ قراءة الصور باستخدام OpenCV
### ما هو الـ Grayscale (التدرج الرمادي)؟
الصورة الملونة تتكون من 3 قنوات: **R (أحمر) + G (أخضر) + B (أزرق)**

الصورة الرمادية تتكون من **قناة واحدة فقط** تمثل شدة الضوء (0 = أسود، 255 = أبيض)

**لماذا نستخدم الرمادي في الفحص الصناعي؟**
- تقليل حجم البيانات بمقدار 3× (أسرع في المعالجة)
- العيوب كالخدوش والشقوق تظهر بوضوح في الرمادي
- الشبكة العصبية تحتاج بيانات أقل لنفس الدقة


````python
# ============================================================
# قراءة الصور وتحويلها إلى تدرج رمادي
# ============================================================
try:
    # cv2.IMREAD_GRAYSCALE = 0 → يقرأ الصورة كرمادية مباشرة
    img_ok  = cv2.imread(path_ok,     cv2.IMREAD_GRAYSCALE)
    img_def = cv2.imread(path_defect, cv2.IMREAD_GRAYSCALE)

    # التحقق من نجاح قراءة الصور
    if img_ok is None:
        raise ValueError(f"فشل تحميل الصورة السليمة: {path_ok}")
    if img_def is None:
        raise ValueError(f"فشل تحميل الصورة المعيبة: {path_defect}")

    print(f"✅ أبعاد صورة القطعة السليمة  : {img_ok.shape}  (الارتفاع × العرض)")
    print(f"✅ أبعاد صورة القطعة المعيبة  : {img_def.shape}  (الارتفاع × العرض)")
    print(f"📊 نطاق قيم البكسل: 0 (أسود) → 255 (أبيض)")
    print(f"📊 متوسط شدة الصورة السليمة  : {img_ok.mean():.1f}")
    print(f"📊 متوسط شدة الصورة المعيبة  : {img_def.mean():.1f}")

except Exception as e:
    print(f"❌ خطأ في تحميل الصورة: {e}")
````

````python
# ============================================================
# عرض الصورتين جنبًا إلى جنب (الاختبار الأولي للكاميرا)
# ============================================================
plt.figure(figsize=(12, 6))  # حجم الشكل بالإنش

# --- الصورة الأولى: القطعة السليمة ---
plt.subplot(1, 2, 1)
plt.title("✅ The Good Part (OK Part)", fontsize=14, fontweight='bold', color='green')
plt.imshow(img_ok, cmap='gray')   # cmap='gray' لعرض الصورة الرمادية بشكل صحيح
plt.axis('off')                    # إخفاء المحاور

# --- الصورة الثانية: القطعة المعيبة ---
plt.subplot(1, 2, 2)
plt.title("❌ The Defective Part (Defective Part)", fontsize=14, fontweight='bold', color='red')
plt.imshow(img_def, cmap='gray')
plt.axis('off')

# عنوان رئيسي للمقارنة
plt.suptitle("🔍 Computer Vision System Testing — Camera Configuration",
             fontsize=16, fontweight='bold', y=1.02)

plt.tight_layout()

# حفظ الصورة تلقائيًا
import pathlib
save_dir = pathlib.Path(r"E:\AI\AI project\code\outputs\plots")
save_dir.mkdir(parents=True, exist_ok=True)
plt.savefig(save_dir / "01_camera_test_comparison.png", dpi=200, bbox_inches='tight')
print("💾 تم حفظ صورة المقارنة في مجلد outputs/plots")

plt.show()
print("\n✅ تم تهيئة نظام الرؤية الحاسوبية بنجاح!")
````

## 📊 تحليل إضافي — مقارنة الهيستوغرام
### ما هو الهيستوغرام (Histogram)؟
الهيستوغرام يُظهر **توزيع قيم شدة البكسل** في الصورة.
- **محور X**: قيمة الشدة (0 إلى 255)
- **محور Y**: عدد البكسلات بهذه الشدة

**كيف نكتشف العيوب من الهيستوغرام؟**
- القطع المعيبة تمتلك توزيعًا أكثر تشتتًا
- العيوب تُظهر قيمًا متطرفة (قمم في الأطراف)


````python
# ============================================================
# تحليل الهيستوغرام للصورتين
# ============================================================
try:
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    # --- الصف الأول: الصور ---
    axes[0, 0].imshow(img_ok,  cmap='gray')
    axes[0, 0].set_title("The intact piece", fontsize=13, color='green', fontweight='bold')
    axes[0, 0].axis('off')

    axes[0, 1].imshow(img_def, cmap='gray')
    axes[0, 1].set_title("The defective piece", fontsize=13, color='red', fontweight='bold')
    axes[0, 1].axis('off')

    # --- الصف الثاني: الهيستوغرامات ---
    axes[1, 0].hist(img_ok.ravel(),  bins=64, color='green',  alpha=0.7, edgecolor='darkgreen')
    axes[1, 0].set_title("Histogram of the Good Part",  fontsize=12)
    axes[1, 0].set_xlabel("Pixel Intensity (0-255)")
    axes[1, 0].set_ylabel("Number of Pixels")
    axes[1, 0].grid(alpha=0.3)
    axes[1, 0].axvline(img_ok.mean(),  color='darkgreen', linestyle='--', label=f'Average = {img_ok.mean():.1f}')
    axes[1, 0].legend()

    axes[1, 1].hist(img_def.ravel(), bins=64, color='red',    alpha=0.7, edgecolor='darkred')
    axes[1, 1].set_title("Histogram of the Defective Part", fontsize=12)
    axes[1, 1].set_xlabel("Pixel Intensity (0-255)")
    axes[1, 1].set_ylabel("Number of Pixels")
    axes[1, 1].grid(alpha=0.3)
    axes[1, 1].axvline(img_def.mean(), color='darkred',   linestyle='--', label=f'Average = {img_def.mean():.1f}')
    axes[1, 1].legend()

    fig.suptitle("Histogram Analysis of Pixel Intensity — Good vs. Defective Parts", fontsize=15, fontweight='bold')
    plt.tight_layout()
    plt.savefig(save_dir / "01_histogram_analysis.png", dpi=200, bbox_inches='tight')
    print("💾 The histogram analysis was saved.")
    plt.show()

except NameError:
    print("⚠️ Please run the previous cells first to load the images")
````

#  Image Processing Pipeline Visualization
---
### هدف هذا النوت بوك
شرح **مسار معالجة الصورة** الذي تمر به قبل دخولها للشبكة العصبية:
1. **الصورة الأصلية** (Original)
2. **تحويل للرمادي** (Grayscale)
3. **تغيير الحجم** لـ 64×64 (Resize)

### 💡 لماذا نُبسّط الصور؟
المعادلة التالية توضح تقليل الحجم:
- **أصلي**: 512×512×3 قناة = **786,432** قيمة لكل صورة
- **بعد التحويل**: 64×64×1 قناة = **4,096** قيمة فقط
- **نسبة التخفيض**: ~192×

## 📦 تجهيز البيانات — Data Preparation
### ما هو Data Augmentation / Transforms؟
قبل إدخال الصور للشبكة، يجب معالجتها بالخطوات التالية:

| الخطوة | الوصف | السبب |
|--------|-------|-------|
| `Grayscale` | تحويل لتدرج رمادي | تقليل الأبعاد |
| `Resize(64,64)` | تغيير الحجم لـ 64×64 | توحيد المدخلات |
| `ToTensor` | تحويل لـ Tensor | مطلوب لـ PyTorch |

**لماذا 64×64 وليس الحجم الأصلي؟**  
توازن بين الدقة وسرعة التدريب. الشبكة تُبقي على الخصائص المهمة.


````python
# ============================================================
# تعريف تحويلات الصور (Image Transforms)
# ============================================================
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),  # تحويل لرمادي (1 قناة بدلاً من 3)
    transforms.Resize((64, 64)),                   # ضغط الصورة لـ 64×64 بكسل
    transforms.ToTensor()                          # تحويل الصورة لـ Tensor + تطبيع 0-1
])
# ملاحظة: ToTensor تُقسم قيم البكسل تلقائيًا على 255
# أي: القيمة 255 (أبيض) → 1.0 ، والقيمة 0 (أسود) → 0.0

# ============================================================
# تحميل بيانات التدريب
# ============================================================
train_dir = r"E:\AI\AI project\code\archive\temp\casting_data\casting_data\train"

try:
    # ImageFolder: يقرأ المجلدات تلقائيًا ويعتبر كل مجلد فئة مختلفة
    # def_front → فئة 0 (معيب)
    # ok_front  → فئة 1 (سليم)
    dataset = datasets.ImageFolder(root=train_dir, transform=transform)

    # DataLoader: يُجهّز البيانات على دفعات (batches) لتغذية الشبكة
    train_loader = DataLoader(
        dataset,
        batch_size=32,   # عدد الصور في كل دفعة (زيادته تُسرّع التدريب لكن يحتاج RAM أكثر)
        shuffle=True     # خلط الصور لتجنب التحيّز
    )

    print("✅ تم تحميل البيانات بنجاح!")
    print(f"📊 إجمالي صور التدريب : {len(dataset)}")
    print(f"🏷️ الفئات             : {dataset.class_to_idx}")
    print(f"🔄 عدد الدفعات (Batches): {len(train_loader)}")
    print(f"🖼️ حجم الدفعة الواحدة  : 32 صورة")

except Exception as e:
    print(f"❌ خطأ في تحميل البيانات: {e}")
````

## 🏗️ بنية الشبكة العصبية التلافيفية (CNN Architecture)
### رحلة البيانات عبر الشبكة:
```
المدخل: صورة [1 × 64 × 64]
    ↓
Conv1 (1→16 فلتر، 3×3) + ReLU → [16 × 64 × 64]
    ↓
MaxPool2d(2×2) → [16 × 32 × 32]
    ↓
Conv2 (16→32 فلتر، 3×3) + ReLU → [32 × 32 × 32]
    ↓
MaxPool2d(2×2) → [32 × 16 × 16]
    ↓
Flatten → [8192] ميزة
    ↓
FC1: 8192 → 128 + ReLU
    ↓
FC2: 128 → 2 (سليم / معيب)
    ↓
المخرج: احتمالية كل فئة
```

### شرح كل طبقة:
- **Conv2d**: تُطبّق فلاتر تكتشف الحواف والأنماط
- **ReLU**: تُزيل القيم السلبية (f(x) = max(0, x))
- **MaxPool2d**: تُصغّر الخريطة بأخذ أعلى قيمة من كل منطقة 2×2
- **Linear (FC)**: يُصنّف الخصائص المستخرجة


````python
# ============================================================
# تعريف بنية الشبكة العصبية التلافيفية (CNN)
# ============================================================
class DefectDetectionCNN(nn.Module):
    def __init__(self):
        super().__init__()

        # ======================================================
        # الكتلة التلافيفية الأولى (Feature Extraction Block 1)
        # ======================================================
        self.conv1 = nn.Conv2d(
            in_channels=1,   # عدد قنوات المدخل (1 لأن الصورة رمادية)
            out_channels=16, # عدد الفلاتر = 16 خريطة ميزات
            kernel_size=3,   # حجم الفلتر 3×3 بكسل
            padding=1        # حشو الحواف للحفاظ على أبعاد الصورة
        )
        self.relu1 = nn.ReLU()       # دالة التفعيل: تُلغي القيم السلبية
        self.pool1 = nn.MaxPool2d(   # تجمّع الميزات وتُقلل الأبعاد بمقدار النصف
            kernel_size=2,
            stride=2
        )

        # ======================================================
        # الكتلة التلافيفية الثانية (Feature Extraction Block 2)
        # ======================================================
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)  # 16 → 32 فلتر (ميزات أعمق)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(2, 2)                # 32×32 → 16×16

        # ======================================================
        # الطبقات المتصلة بالكامل (Fully Connected Layers)
        # ======================================================
        # بعد التجميع: 32 فلتر × 16 × 16 = 8192 ميزة
        self.fc1 = nn.Linear(32 * 16 * 16, 128)  # تقليل من 8192 → 128 ميزة مضغوطة
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128, 2)             # 128 → 2 (سليم أو معيب)

    def forward(self, x):
        # تسلسل العمليات عبر الشبكة
        x = self.pool1(self.relu1(self.conv1(x)))  # كتلة 1
        x = self.pool2(self.relu2(self.conv2(x)))  # كتلة 2
        x = x.view(x.size(0), -1)  # فرد البيانات (Flatten): [batch, 32, 16, 16] → [batch, 8192]
        x = self.relu3(self.fc1(x))               # طبقة مخفية
        x = self.fc2(x)                            # طبقة الإخراج
        return x

# إنشاء الشبكة
model = DefectDetectionCNN()

# حساب إجمالي المعاملات القابلة للتعلم
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print("✅ تم بناء الشبكة العصبية بنجاح!")
print(f"🔢 إجمالي المعاملات القابلة للتعلم: {total_params:,}")
print(f"\n{model}")
````

## ⚙️ إعداد التدريب
### دالة الخسارة — Cross Entropy Loss
تُقيس **مدى خطأ** تنبؤات الشبكة:
- إذا كانت النبوءة صحيحة بثقة عالية → خسارة منخفضة
- إذا كانت النبوءة خاطئة → خسارة عالية

### خوارزمية التحسين — Adam Optimizer
تُحدّث أوزان الشبكة بعد كل دفعة لتقليل الخسارة.  
**معدل التعلم `lr=0.001`**:
- قيمة كبيرة → تعلم سريع لكن غير مستقر
- قيمة صغيرة → تعلم دقيق لكن بطيء
- **0.001 قيمة مثالية للبداية**


````python
# ============================================================
# إعداد دالة الخسارة والمُحسِّن
# ============================================================
criterion = nn.CrossEntropyLoss()   # دالة الخسارة للتصنيف المتعدد الفئات
optimizer = optim.Adam(
    model.parameters(),  # المعاملات التي سيُحدّثها Adam
    lr=0.001             # معدل التعلم (Learning Rate)
)

epochs = 3  # عدد مرات المرور الكامل على بيانات التدريب
# ملاحظة: 3 epochs سريعة للتجربة، يُنصح بـ 10-50 epoch للنتائج المثلى

print(f"⚙️  دالة الخسارة   : {criterion.__class__.__name__}")
print(f"⚙️  خوارزمية التحسين: Adam")
print(f"⚙️  معدل التعلم    : 0.001")
print(f"⚙️  عدد الـ Epochs  : {epochs}")
````

````python
# ============================================================
# حلقة التدريب الرئيسية الخاصة بشبكة CNN
# Main Training Loop for CNN Model
# ============================================================

# إنشاء قائمة فارغة لتخزين نتائج كل Epoch
# سيتم حفظ الـ Loss والـ Accuracy داخلها لاستخدامها لاحقًا في الرسومات البيانية
history = []

# طباعة رسالة بداية التدريب
print("🚀 بدء التدريب...\n")

# طباعة عناوين الأعمدة أثناء عرض النتائج
# Epoch = رقم الدورة التدريبية
# Loss = نسبة الخطأ
# Accuracy = نسبة الدقة
print(f"{'Epoch':<8} {'الخسارة (Loss)':<20} {'الدقة (Accuracy)':<20}")

# طباعة خط فاصل للتنسيق
print("=" * 50)

# ============================================================
# محاولة تشغيل التدريب مع التقاط أي أخطاء محتملة
# ============================================================

try:

    # ========================================================
    # حلقة الـ Epochs
    # ========================================================

    # الشبكة ستكرر التدريب بعدد الـ Epochs المحدد
    # كل Epoch = مرور كامل على جميع صور التدريب
    for epoch in range(epochs):

        # ====================================================
        # متغيرات حساب الأداء داخل الـ Epoch الحالية
        # ====================================================

        # running_loss:
        # يجمع إجمالي قيمة الخطأ أثناء الـ Epoch
        running_loss = 0.0

        # correct:
        # يحسب عدد الصور التي تم تصنيفها بشكل صحيح
        correct = 0

        # total:
        # يحسب العدد الكلي للصور التي تم معالجتها
        total = 0

        # ====================================================
        # تفعيل وضع التدريب
        # ====================================================

        # model.train():
        # يضع الشبكة في وضع التدريب
        # وهذا يُفعّل طبقات مثل:
        # Dropout و BatchNorm بطريقة التدريب الصحيحة
        model.train()

        # ====================================================
        # حلقة قراءة الـ Batches من الـ DataLoader
        # ====================================================

        # train_loader يقوم بإرسال:
        # images = صور التدريب
        # labels = التصنيفات الحقيقية للصور
        for images, labels in train_loader:

            # =================================================
            # الخطوة 1 — تصفير الـ Gradients القديمة
            # =================================================

            # PyTorch يخزن الـ gradients تلقائيًا
            # لذلك يجب تصفيرها قبل كل Batch جديد
            optimizer.zero_grad()

            # =================================================
            # الخطوة 2 — التمرير الأمامي Forward Pass
            # =================================================

            # تمرير الصور داخل الشبكة العصبية CNN
            # الشبكة تقوم بـ:
            # - استخراج الخصائص Features
            # - تحليل الصورة
            # - إنتاج احتمالات التصنيف
            outputs = model(images)

            # =================================================
            # الخطوة 3 — حساب الخطأ Loss
            # =================================================

            # مقارنة توقعات الشبكة بالإجابات الحقيقية
            # CrossEntropyLoss تحسب مقدار الخطأ
            loss = criterion(outputs, labels)

            # =================================================
            # الخطوة 4 — Backpropagation
            # =================================================

            # حساب الـ gradients لكل وزن داخل الشبكة
            # باستخدام خوارزمية الانتشار الخلفي
            loss.backward()

            # =================================================
            # الخطوة 5 — تحديث الأوزان Weights Update
            # =================================================

            # Adam Optimizer يستخدم الـ gradients
            # لتعديل أوزان الشبكة وتقليل الخطأ
            optimizer.step()

            # =================================================
            # حساب إجمالي الـ Loss
            # =================================================

            # loss.item():
            # يحول قيمة الـ tensor إلى رقم عادي
            running_loss += loss.item()

            # =================================================
            # استخراج التوقعات النهائية
            # =================================================

            # outputs.data يحتوي احتمالات التصنيف
            # torch.max يختار أعلى احتمال
            #
            # مثال:
            # [0.1 , 0.9]
            #
            # سيتم اختيار الفئة رقم 1
            _, predicted = torch.max(outputs.data, 1)

            # =================================================
            # تحديث عدد الصور الكلي
            # =================================================

            # labels.size(0):
            # عدد الصور داخل الـ Batch الحالية
            total += labels.size(0)

            # =================================================
            # حساب عدد التوقعات الصحيحة
            # =================================================

            # predicted == labels:
            # مقارنة توقعات الشبكة بالإجابات الحقيقية
            #
            # sum():
            # حساب عدد القيم الصحيحة True
            #
            # item():
            # تحويل الناتج إلى رقم عادي
            correct += (predicted == labels).sum().item()

        # ====================================================
        # حساب متوسط الـ Loss للـ Epoch كاملة
        # ====================================================

        # running_loss:
        # إجمالي الخطأ لجميع الـ batches
        #
        # len(train_loader):
        # عدد الـ batches
        #
        # الناتج:
        # متوسط الخطأ الكلي
        epoch_loss = running_loss / max(1, len(train_loader))

        # ====================================================
        # حساب الـ Accuracy
        # ====================================================

        # معادلة الدقة:
        #
        # Accuracy = Correct Predictions / Total Samples × 100
        #
        # correct:
        # عدد الصور المصنفة بشكل صحيح
        #
        # total:
        # إجمالي الصور
        epoch_acc = 100 * correct / max(1, total)

        # ====================================================
        # حفظ نتائج الـ Epoch الحالية
        # ====================================================

        # حفظ:
        # - رقم الـ Epoch
        # - الـ Loss
        # - الـ Accuracy
        #
        # لاستخدامها لاحقًا في:
        # - الرسومات البيانية
        # - تحليل الأداء
        history.append({

            # رقم الـ Epoch الحالية
            'epoch': epoch + 1,

            # قيمة الخطأ
            'loss': epoch_loss,

            # نسبة الدقة
            'accuracy': epoch_acc
        })

        # ====================================================
        # طباعة نتائج التدريب الحالية
        # ====================================================

        print(
            f"  {epoch+1}/{epochs}    "
            f"{epoch_loss:.4f}              "
            f"{epoch_acc:.2f}%"
        )

    # ========================================================
    # رسالة انتهاء التدريب
    # ========================================================

    print("\n🎉 اكتمل التدريب بنجاح!")

# ============================================================
# معالجة الأخطاء إن وجدت
# ============================================================

except Exception as e:

    # طباعة رسالة الخطأ
    print(f"❌ خطأ أثناء التدريب: {e}")

    # اقتراح سبب المشكلة
    print("💡 تأكد من تحميل البيانات في الخلايا السابقة")
````

### ما هو نظام الخبير (Expert System)؟
نظام الخبير هو **برنامج حاسوبي** يُحاكي قرارات الخبير البشري باستخدام قواعد منطقية.

**في مشروعنا:**
- **المدخل**: نتيجة الشبكة العصبية (سليم أو معيب)
- **القواعد**: إذا معيب → أرسل للسلة / إذا سليم → أرسل للتغليف
- **المخرج**: إحداثيات ذراع الروبوت (X, Y, Z)


````python
# ============================================================
# استيراد المكتبات
# ============================================================
import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image
import os, random
from pathlib import Path

print("✅ تم تحميل المكتبات")
````

## 🏗️ تعريف نموذج CNN (نفس البنية المدرّبة)
يجب أن تكون بنية الشبكة **مطابقة تمامًا** للبنية المستخدمة أثناء التدريب،
وإلا لن تتمكن من تحميل الأوزان المحفوظة.


````python
# ============================================================
# تعريف نموذج الشبكة العصبية CNN الخاصة باكتشاف العيوب
# CNN Architecture for Industrial Defect Detection
# ============================================================

# إنشاء Class جديدة ترث من nn.Module
# nn.Module هو الكلاس الأساسي لأي شبكة عصبية في PyTorch
class DefectDetectionCNN(nn.Module):

    # ========================================================
    # دالة البناء Constructor
    # ========================================================

    # يتم تنفيذها مرة واحدة عند إنشاء النموذج
    def __init__(self):

        # استدعاء الـ constructor الأساسي من nn.Module
        # ضروري جدًا لكي تعمل الشبكة بشكل صحيح
        super().__init__()

        # ====================================================
        # الطبقة التلافيفية الأولى Convolution Layer 1
        # ====================================================

        # nn.Conv2d:
        # طبقة استخراج الخصائص Features من الصور
        #
        # المعاملات:
        #
        # 1  → عدد قنوات الإدخال
        #      لأن الصور Grayscale
        #
        # 16 → عدد الفلاتر Filters
        #      أي سيتم استخراج 16 Feature Map مختلفة
        #
        # 3  → حجم الـ Kernel
        #      أي نافذة 3×3 تتحرك فوق الصورة
        #
        # padding=1:
        # إضافة إطار حول الصورة للحفاظ على الأبعاد
        #
        # هذه الطبقة تتعلم:
        # - الحواف Edges
        # - الخطوط
        # - التفاصيل الصناعية
        self.conv1 = nn.Conv2d(1, 16, 3, padding=1)

        # ====================================================
        # دالة التفعيل ReLU الأولى
        # ====================================================

        # ReLU:
        # Rectified Linear Unit
        #
        # وظيفتها:
        # - إدخال Non-Linearity للشبكة
        # - إزالة القيم السالبة
        #
        # المعادلة:
        #
        # f(x) = max(0, x)
        #
        # إذا كانت القيمة سالبة:
        # تتحول إلى صفر
        #
        # إذا كانت موجبة:
        # تبقى كما هي
        #
        # تساعد الشبكة على تعلم الأنماط المعقدة
        self.relu1 = nn.ReLU()

        # ====================================================
        # طبقة الـ Max Pooling الأولى
        # ====================================================

        # MaxPool2d:
        # تقوم بتقليل أبعاد الصورة
        #
        # (2,2):
        # نافذة 2×2
        #
        # تختار أكبر قيمة فقط
        #
        # الفوائد:
        # - تقليل حجم البيانات
        # - تقليل الحسابات
        # - الاحتفاظ بأهم الخصائص
        #
        # مثال:
        #
        # [1 5]
        # [3 2]
        #
        # الناتج = 5
        self.pool1 = nn.MaxPool2d(2, 2)

        # ====================================================
        # الطبقة التلافيفية الثانية
        # ====================================================

        # هنا تدخل 16 Feature Maps
        # وتخرج 32 Feature Maps
        #
        # الطبقة الثانية تتعلم خصائص أعمق مثل:
        # - الشقوق
        # - التشوهات
        # - العيوب الصناعية المعقدة
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)

        # دالة ReLU الثانية
        self.relu2 = nn.ReLU()

        # طبقة الـ Pooling الثانية
        self.pool2 = nn.MaxPool2d(2, 2)

        # ====================================================
        # الطبقة المتصلة Fully Connected Layer 1
        # ====================================================

        # بعد انتهاء الـ CNN Layers
        # يتم تحويل البيانات إلى Vector
        #
        # 32 * 16 * 16:
        # الحجم النهائي الخارج من الـ CNN
        #
        # لماذا؟
        #
        # الصورة الأصلية:
        # 64×64
        #
        # بعد Pooling أول:
        # 32×32
        #
        # بعد Pooling ثاني:
        # 16×16
        #
        # وعدد الـ Channels = 32
        #
        # إذًا:
        #
        # 32 × 16 × 16 = 8192
        #
        # هذه الطبقة تبدأ مرحلة التصنيف Classification
        self.fc1 = nn.Linear(32 * 16 * 16, 128)

        # دالة ReLU الثالثة
        self.relu3 = nn.ReLU()

        # ====================================================
        # طبقة الإخراج النهائية Output Layer
        # ====================================================

        # 128:
        # عدد الـ neurons القادمة من الطبقة السابقة
        #
        # 2:
        # عدد الفئات Classes
        #
        # في المشروع:
        # - Defective
        # - Non-Defective
        #
        # هذه الطبقة تنتج الاحتمالات النهائية للتصنيف
        self.fc2 = nn.Linear(128, 2)

    # ========================================================
    # دالة الـ Forward Pass
    # ========================================================

    # هذه الدالة تحدد مسار مرور البيانات داخل الشبكة
    def forward(self, x):

        # ====================================================
        # المرحلة الأولى من استخراج الخصائص
        # ====================================================

        # conv1(x):
        # تطبيق الـ Convolution الأولى
        #
        # relu1():
        # تطبيق دالة التفعيل
        #
        # pool1():
        # تقليل الأبعاد
        x = self.pool1(self.relu1(self.conv1(x)))

        # ====================================================
        # المرحلة الثانية من استخراج الخصائص
        # ====================================================

        # الطبقة الثانية تتعلم خصائص أكثر تعقيدًا
        x = self.pool2(self.relu2(self.conv2(x)))

        # ====================================================
        # Flatten Operation
        # ====================================================

        # تحويل الـ Feature Maps من شكل 3D
        # إلى Vector واحد
        #
        # x.size(0):
        # عدد الصور داخل الـ Batch
        #
        # -1:
        # PyTorch يحسب الحجم تلقائيًا
        #
        # مثال:
        #
        # [32, 32, 16, 16]
        #
        # تصبح:
        #
        # [32, 8192]
        x = x.view(x.size(0), -1)

        # ====================================================
        # الطبقة المتصلة الأولى
        # ====================================================

        # Fully Connected Layer
        #
        # تبدأ اتخاذ قرار التصنيف
        x = self.relu3(self.fc1(x))

        # ====================================================
        # طبقة الإخراج النهائية
        # ====================================================

        # تنتج الاحتمالات النهائية للفئات
        #
        # مثال:
        #
        # [0.12 , 0.88]
        #
        # أي:
        # 12% Defective
        # 88% OK
        return self.fc2(x)
````

## 📋 قواعد نظام الخبير
```
إذا (الفئة == 0) أي معيب:
    تفعيل القاعدة: "IF Defect THEN Move to Scrap Bin"
    إحداثيات الروبوت: X=250, Y=-150, Z=100
    القرار: REJECT (رفض)

إذا (الفئة == 1) أي سليم:
    تفعيل القاعدة: "IF OK THEN Move to Packaging Line"
    إحداثيات الروبوت: X=250, Y=+150, Z=100
    القرار: ACCEPT (قبول)
```


````python
# ============================================================
# دالة نظام الخبير لاتخاذ القرار الصناعي
# Expert System Decision Making Function
# ============================================================

# تعريف دالة جديدة اسمها expert_system_decision
#
# وظيفتها:
# تحويل نتيجة الذكاء الاصطناعي (CNN)
# إلى أوامر حركة فعلية للروبوت
#
# المدخل:
# prediction_class
#
# 0 = Defective
# 1 = OK
#
# المخرج:
# إحداثيات حركة الروبوت [X, Y, Z]
def expert_system_decision(prediction_class):

    # ========================================================
    # شرح الدالة باستخدام Docstring
    # ========================================================

    """
    Converts CNN prediction results into robotic commands.

    Input:
        prediction_class
            0 = Defective Part
            1 = Non-Defective Part

    Output:
        Robot target coordinates [X, Y, Z]
    """

    # ========================================================
    # طباعة رسالة بدء نظام الخبير
    # ========================================================

    print("\n[ Expert System Processing... ]")

    # ========================================================
    # الحالة الأولى — القطعة معيبة
    # ========================================================

    # إذا كانت نتيجة الـ CNN = 0
    #
    # فهذا يعني:
    # القطعة تحتوي على عيب صناعي
    if prediction_class == 0:

        # ====================================================
        # القاعدة المنطقية Rule Base
        # ====================================================

        # قاعدة نظام الخبير:
        #
        # IF Defect
        # THEN Move to Scrap Bin
        #
        # أي:
        # إذا كانت القطعة معيبة
        # انقلها إلى سلة الخردة
        rule_triggered = "IF Defect THEN Move to Scrap Bin"

        # ====================================================
        # تحديد إحداثيات الروبوت
        # ====================================================

        # الإحداثيات المطلوبة لحركة الروبوت
        #
        # X = 250 mm
        # Y = -150 mm
        # Z = 100 mm
        #
        # الاتجاه السالب في محور Y
        # يعني التحرك ناحية اليسار
        #
        # وهو مكان صندوق الخردة
        target_coordinates = [250.0, -150.0, 100.0]

        # ====================================================
        # تحديد الإجراء النهائي
        # ====================================================

        # الإجراء:
        # رفض القطعة
        action = "REJECT → Move to Scrap Bin"

    # ========================================================
    # الحالة الثانية — القطعة سليمة
    # ========================================================

    # إذا كانت نتيجة الـ CNN = 1
    #
    # فهذا يعني:
    # القطعة سليمة
    elif prediction_class == 1:

        # ====================================================
        # قاعدة نظام الخبير
        # ====================================================

        # IF OK
        # THEN Move to Packaging Line
        #
        # أي:
        # إذا كانت القطعة سليمة
        # يتم إرسالها إلى خط التغليف
        rule_triggered = "IF OK THEN Move to Packaging Line"

        # ====================================================
        # تحديد الإحداثيات
        # ====================================================

        # Y موجب
        #
        # أي التحرك ناحية اليمين
        #
        # حيث خط التغليف
        target_coordinates = [250.0, 150.0, 100.0]

        # ====================================================
        # الإجراء النهائي
        # ====================================================

        action = "ACCEPT → Move to Packaging Line"

    # ========================================================
    # الحالة الثالثة — خطأ غير معروف
    # ========================================================

    # إذا وصلت قيمة غير 0 أو 1
    #
    # فهذا يعني:
    # يوجد خطأ في النظام
    else:

        # ====================================================
        # قاعدة الطوارئ
        # ====================================================

        # إيقاف النظام لحماية الروبوت
        rule_triggered = "UNKNOWN ERROR → Emergency Stop"

        # ====================================================
        # وضع الروبوت في موضع آمن
        # ====================================================

        # رفع الروبوت لأعلى
        #
        # Z = 500 mm
        #
        # لمنع الاصطدام
        target_coordinates = [0.0, 0.0, 500.0]

        # ====================================================
        # تنفيذ إيقاف الطوارئ
        # ====================================================

        action = "EMERGENCY STOP"

    # ========================================================
    # عرض نتائج نظام الخبير
    # ========================================================

    # طباعة القاعدة التي تم تنفيذها
    print(f"Activated Rule : {rule_triggered}")

    # طباعة الإجراء النهائي
    print(f"Robot Action   : {action}")

    # طباعة الإحداثيات المطلوبة
    #
    # target_coordinates[0] = X
    # target_coordinates[1] = Y
    # target_coordinates[2] = Z
    print(
        f"Coordinates (mm): "
        f"X={target_coordinates[0]}, "
        f"Y={target_coordinates[1]}, "
        f"Z={target_coordinates[2]}"
    )

    # ========================================================
    # إعادة الإحداثيات النهائية
    # ========================================================

    # return:
    # ترجع الإحداثيات إلى البرنامج الرئيسي
    #
    # لكي يستخدمها الروبوت أو الـ Digital Twin
    return target_coordinates

# ============================================================
# رسالة تأكيد تعريف الدالة
# ============================================================

print("Expert system function defined successfully.")
````

````python
# ============================================================
# تشغيل نظام الفحص الكامل
# Full Inspection Pipeline
#
# Vision AI + Expert System + Robot Decision Making
# ============================================================

# ============================================================
# استيراد المكتبات المطلوبة
# ============================================================

# matplotlib:
# تستخدم لرسم النتائج والـ Graphs
import matplotlib.pyplot as plt

# numpy:
# تستخدم لمعالجة الصور والمصفوفات
import numpy as np

# ============================================================
# التأكد من وجود الصورة والموديل
# ============================================================

# image_path:
# مسار صورة الاختبار
#
# MODEL_PTH.exists():
# التأكد من وجود ملف النموذج المدرب
#
# لن يبدأ النظام إذا لم تكن الصورة أو الموديل موجودين
if image_path and MODEL_PTH.exists():

    # ========================================================
    # الخطوة 1 — تحميل صورة الاختبار
    # ========================================================

    # Image.open():
    # فتح الصورة باستخدام مكتبة PIL
    #
    # الصورة هنا تمثل قطعة صناعية جديدة
    # سيتم فحصها بواسطة الذكاء الاصطناعي
    image = Image.open(image_path)

    # ========================================================
    # الخطوة 2 — تجهيز الصورة للشبكة العصبية
    # ========================================================

    # transform(image):
    # تطبيق الـ preprocessing على الصورة
    #
    # مثل:
    # - Resize
    # - Grayscale
    # - Tensor Conversion
    # - Normalization
    #
    # unsqueeze(0):
    # إضافة Batch Dimension
    #
    # لأن PyTorch يتوقع:
    #
    # [Batch, Channel, Height, Width]
    #
    # فتتحول الصورة من:
    #
    # [1,64,64]
    #
    # إلى:
    #
    # [1,1,64,64]
    input_tensor = transform(image).unsqueeze(0)

    # ========================================================
    # الخطوة 3 — تشغيل الشبكة العصبية CNN
    # ========================================================

    # torch.no_grad():
    #
    # تعطيل حساب الـ gradients
    #
    # لأننا هنا في مرحلة Evaluation
    # وليس Training
    #
    # هذا:
    # - يقلل استهلاك الذاكرة
    # - يزيد السرعة
    with torch.no_grad():

        # ====================================================
        # إدخال الصورة إلى الشبكة
        # ====================================================

        # model(input_tensor):
        # تنفيذ الـ Forward Pass
        #
        # الشبكة تقوم بتحليل الصورة
        # واستخراج الخصائص الصناعية
        output = model(input_tensor)

        # ====================================================
        # تحويل المخرجات إلى احتمالات
        # ====================================================

        # softmax():
        #
        # تحول القيم الخام Raw Outputs
        # إلى احتمالات حقيقية
        #
        # مجموعها = 100%
        #
        # مثال:
        #
        # [2.5 , 0.3]
        #
        # تصبح:
        #
        # [0.90 , 0.10]
        probabilities = torch.softmax(output, dim=1)

        # ====================================================
        # استخراج أعلى احتمال
        # ====================================================

        # torch.max():
        #
        # اختيار أعلى Probability
        #
        # predicted:
        # رقم الفئة المتوقعة
        #
        # مثال:
        #
        # 0 = Defective
        # 1 = OK
        _, predicted = torch.max(probabilities, 1)

        # ====================================================
        # تحويل Tensor إلى رقم عادي
        # ====================================================

        # item():
        # استخراج القيمة الرقمية
        predicted_class = predicted.item()

        # ====================================================
        # حساب نسبة الثقة
        # ====================================================

        # probabilities[0][predicted_class]:
        #
        # استخراج احتمال الفئة المتوقعة
        #
        # *100:
        # تحويله إلى نسبة مئوية
        confidence = probabilities[0][predicted_class].item() * 100

    # ========================================================
    # الخطوة 4 — تفسير نتيجة الـ CNN
    # ========================================================

    # إذا كانت الفئة = 0
    # فهذا يعني أن القطعة معيبة
    #
    # وإلا فهي سليمة
    vision_result = (
        "DEFECTIVE"
        if predicted_class == 0
        else "OK"
    )

    # ========================================================
    # استخراج احتمالات جميع الفئات
    # ========================================================

    # class_probs:
    # Dictionary يحتوي على احتمالات:
    # - Defective
    # - OK
    class_probs = {

        # احتمال أن تكون القطعة معيبة
        'Defective': probabilities[0][0].item() * 100,

        # احتمال أن تكون القطعة سليمة
        'OK': probabilities[0][1].item() * 100,
    }

    # ========================================================
    # عرض نتائج الذكاء الاصطناعي
    # ========================================================

    print(f"\n{'='*45}")

    # عرض النتيجة النهائية
    print(f"AI Vision Result : {vision_result}")

    # عرض نسبة الثقة
    print(f"Decision Confidence : {confidence:.1f}%")

    # عرض احتمال العيب
    print(f"Defective Probability : {class_probs['Defective']:.1f}%")

    # عرض احتمال السلامة
    print(f"OK Probability : {class_probs['OK']:.1f}%")

    print(f"{'='*45}")

    # ========================================================
    # الخطوة 5 — تشغيل نظام الخبير
    # ========================================================

    # إرسال نتيجة الـ CNN إلى نظام الخبير
    #
    # ليقوم بتحويل القرار
    # إلى أوامر حركة للروبوت
    robot_coords = expert_system_decision(predicted_class)

    # ========================================================
    # الخطوة 6 — عرض النتائج بصريًا
    # ========================================================

    # إنشاء نافذة تحتوي على رسمين
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # ========================================================
    # تجهيز الصورة للعرض
    # ========================================================

    # convert('L'):
    # تحويل الصورة إلى Grayscale
    img_display = np.array(image.convert('L'))

    # ========================================================
    # تحديد لون العرض حسب النتيجة
    # ========================================================

    # إذا كانت القطعة معيبة:
    # استخدم اللون الأحمر
    #
    # إذا كانت سليمة:
    # استخدم اللون الأخضر
    color = 'Reds' if predicted_class == 0 else 'Greens'

    # عرض الصورة
    ax1.imshow(img_display, cmap=color)

    # ========================================================
    # تحديد لون عنوان الصورة
    # ========================================================

    title_color = 'red' if predicted_class == 0 else 'green'

    # ========================================================
    # كتابة نتيجة الفحص فوق الصورة
    # ========================================================

    ax1.set_title(
        f"{vision_result}\nConfidence: {confidence:.1f}%",
        fontsize=13,
        color=title_color,
        fontweight='bold'
    )

    # إخفاء المحاور
    ax1.axis('off')

    # ========================================================
    # رسم مخطط الاحتمالات
    # ========================================================

    # barh():
    # رسم Bar Chart أفقي
    bars = ax2.barh(

        # أسماء الفئات
        ['Defective', 'OK'],

        # قيم الاحتمالات
        [
            class_probs['Defective'],
            class_probs['OK']
        ],

        # ألوان الأعمدة
        color=['#e74c3c', '#2ecc71']
    )

    # ========================================================
    # إعدادات الرسم البياني
    # ========================================================

    # تحديد مدى المحور الأفقي
    ax2.set_xlim(0, 100)

    # اسم المحور الأفقي
    ax2.set_xlabel("Probability (%)")

    # عنوان الرسم
    ax2.set_title(
        "Probability Distribution",
        fontsize=13,
        fontweight='bold'
    )

    # إضافة Grid
    ax2.grid(axis='x', alpha=0.3)

    # ========================================================
    # كتابة النسب فوق الأعمدة
    # ========================================================

    for bar, val in zip(
        bars,
        [
            class_probs['Defective'],
            class_probs['OK']
        ]
    ):

        # text():
        # كتابة النص فوق العمود
        ax2.text(

            # موضع الكتابة X
            bar.get_width() + 1,

            # موضع الكتابة Y
            bar.get_y() + bar.get_height()/2,

            # النص المعروض
            f'{val:.1f}%',

            # محاذاة رأسية
            va='center',

            # حجم الخط
            fontsize=11
        )

    # ========================================================
    # العنوان الرئيسي للشكل بالكامل
    # ========================================================

    fig.suptitle(
        "Expert System Inspection Results",
        fontsize=15,
        fontweight='bold'
    )

    # ضبط المسافات
    plt.tight_layout()

    # ========================================================
    # حفظ النتائج
    # ========================================================

    # تحديد فولدر الحفظ
    save_dir = Path(r"E:\AI\AI project\code\outputs\plots")

    # إنشاء الفولدر إذا لم يكن موجودًا
    save_dir.mkdir(parents=True, exist_ok=True)

    # حفظ الشكل كصورة
    plt.savefig(
        save_dir / "03_expert_system_result.png",
        dpi=200,
        bbox_inches='tight'
    )

    # عرض النتائج
    plt.show()

    # ========================================================
    # عرض الإحداثيات النهائية للروبوت
    # ========================================================

    print(
        f"\nRobot Coordinates Sent: "
        f"X={robot_coords[0]}, "
        f"Y={robot_coords[1]}, "
        f"Z={robot_coords[2]} mm"
    )

    # ========================================================
    # رسالة جاهزية النظام
    # ========================================================

    print("[ System Ready: Waiting for MATLAB Transmission ]")

# ============================================================
# في حالة عدم وجود البيانات
# ============================================================

else:

    # رسالة خطأ
    print(
        "Please make sure the trained model "
        "and test image are available."
    )
````

````python
# ============================================================
# تعريف وتحميل النموذج
# ============================================================
class DefectDetectionCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(2, 2)
        self.fc1   = nn.Linear(32 * 16 * 16, 128)
        self.relu3 = nn.ReLU()
        self.fc2   = nn.Linear(128, 2)

    def forward(self, x):
        x = self.pool1(self.relu1(self.conv1(x)))
        x = self.pool2(self.relu2(self.conv2(x)))
        x = x.view(x.size(0), -1)
        x = self.relu3(self.fc1(x))
        return self.fc2(x)

MODEL_PTH = Path(r"E:\AI\AI project\code\data\robot_vision_brain.pth")
model = DefectDetectionCNN()

if MODEL_PTH.exists():
    model.load_state_dict(torch.load(MODEL_PTH, map_location='cpu'))
    model.eval()
    print("✅ تم تحميل النموذج بنجاح")
else:
    print("⚠️ النموذج غير موجود — يرجى تشغيل النوت بوك الثاني أولاً")
````

````python
# ============================================================
# تشغيل تقييم النموذج على بيانات الاختبار
# Model Evaluation on Test Dataset
# ============================================================

try:

    # ========================================================
    # إنشاء قوائم لتخزين النتائج
    # ========================================================

    # all_predictions:
    # ستحتوي على جميع توقعات النموذج
    #
    # مثال:
    # [0,1,1,0,1]
    all_predictions = []

    # all_labels:
    # ستحتوي على التصنيفات الحقيقية
    #
    # مثال:
    # [0,1,0,0,1]
    all_labels = []

    # ========================================================
    # متغيرات حساب الدقة
    # ========================================================

    # correct:
    # عدد التوقعات الصحيحة
    correct = 0

    # total:
    # إجمالي عدد الصور
    total = 0

    # ========================================================
    # تعطيل حساب الـ Gradients
    # ========================================================

    # torch.no_grad():
    #
    # نستخدمه أثناء التقييم فقط
    #
    # لأنه:
    # - يقلل استهلاك الذاكرة
    # - يزيد سرعة التنفيذ
    # - لا نحتاج Backpropagation هنا
    with torch.no_grad():

        # ====================================================
        # المرور على جميع صور الاختبار
        # ====================================================

        # test_loader:
        # يرسل الصور على شكل Batches
        #
        # images:
        # صور الاختبار
        #
        # labels:
        # التصنيفات الحقيقية
        for images, labels in test_loader:

            # =================================================
            # تمرير الصور داخل الشبكة العصبية
            # =================================================

            # model(images):
            # تنفيذ الـ Forward Pass
            #
            # الشبكة تحلل الصور
            # وتنتج احتمالات التصنيف
            outputs = model(images)

            # =================================================
            # استخراج أعلى احتمال
            # =================================================

            # torch.max():
            #
            # اختيار أعلى Probability
            #
            # outputs.data:
            # القيم الناتجة من الشبكة
            #
            # dim=1:
            # البحث داخل الفئات Classes
            #
            # predicted:
            # الفئة المتوقعة لكل صورة
            _, predicted = torch.max(outputs.data, 1)

            # =================================================
            # تحديث إجمالي عدد الصور
            # =================================================

            # labels.size(0):
            # عدد الصور داخل الـ Batch الحالية
            total += labels.size(0)

            # =================================================
            # حساب عدد التوقعات الصحيحة
            # =================================================

            # predicted == labels:
            # مقارنة توقعات الشبكة بالحقيقة
            #
            # sum():
            # حساب عدد النتائج الصحيحة
            #
            # item():
            # تحويل القيمة إلى رقم عادي
            correct += (predicted == labels).sum().item()

            # =================================================
            # حفظ التوقعات داخل القائمة
            # =================================================

            # predicted.cpu():
            # نقل البيانات من GPU إلى CPU
            #
            # numpy():
            # تحويل Tensor إلى NumPy Array
            #
            # tolist():
            # تحويلها إلى List عادية
            all_predictions.extend(
                predicted.cpu().numpy().tolist()
            )

            # =================================================
            # حفظ الـ Labels الحقيقية
            # =================================================

            all_labels.extend(
                labels.cpu().numpy().tolist()
            )

    # ========================================================
    # حساب دقة النموذج
    # ========================================================

    # معادلة الـ Accuracy:
    #
    # Accuracy =
    # (عدد التوقعات الصحيحة / إجمالي العينات) × 100
    #
    # مثال:
    #
    # correct = 95
    # total   = 100
    #
    # accuracy = 95%
    accuracy = 100 * correct / total

    # ========================================================
    # عرض الدقة النهائية
    # ========================================================

    print("=" * 45)

    print(
        f"Model Accuracy on Test Dataset = "
        f"{accuracy:.2f}%"
    )

    print("=" * 45)

    # ========================================================
    # استخراج أسماء الفئات
    # ========================================================

    # class_to_idx:
    # Dictionary داخل Dataset
    #
    # مثال:
    #
    # {
    #   'Defective':0,
    #   'OK':1
    # }
    #
    # keys():
    # استخراج أسماء الفئات فقط
    target_names = list(
        test_dataset.class_to_idx.keys()
    )

    # ========================================================
    # عرض تقرير التصنيف الكامل
    # ========================================================

    print("\nDetailed Classification Report:")

    # classification_report():
    #
    # دالة من sklearn
    #
    # تعرض:
    #
    # - Precision
    # - Recall
    # - F1-Score
    # - Support
    #
    # all_labels:
    # التصنيفات الحقيقية
    #
    # all_predictions:
    # توقعات النموذج
    #
    # target_names:
    # أسماء الفئات
    print(
        classification_report(
            all_labels,
            all_predictions,
            target_names=target_names
        )
    )

# ============================================================
# معالجة الأخطاء
# ============================================================

except Exception as e:

    # طباعة الخطأ إذا حدثت مشكلة
    print(f"Error: {e}")
````

## 🔲 Confusion Matrix
```
                  التنبؤ: معيبة    التنبؤ: سليمة
الحقيقة: معيبة  [  TP (صح)    |   FN (خطأ)   ]
الحقيقة: سليمة  [  FP (خطأ)   |   TN (صح)    ]
```
- **TP (True Positive)**: معيبة صُنّفت معيبة ✅
- **TN (True Negative)**: سليمة صُنّفت سليمة ✅  
- **FP (False Positive)**: سليمة صُنّفت معيبة ⚠️
- **FN (False Negative)**: معيبة صُنّفت سليمة ❌ (الأخطر!)


````python
# ============================================================
# رسم مصفوفة الارتباك الخاصة بتقييم النموذج
# Confusion Matrix Visualization
# ============================================================

try:

    # ========================================================
    # إنشاء مسار حفظ الرسومات
    # ========================================================

    # Path():
    # تحديد مكان حفظ الصور والرسومات
    PLOTS_DIR = Path(r"E:\AI\AI project\code\outputs\plots")

    # mkdir():
    # إنشاء الفولدر تلقائيًا إذا لم يكن موجودًا
    PLOTS_DIR.mkdir(parents=True, exist_ok=True)

    # ========================================================
    # تحديد اسم ملف الصورة
    # ========================================================

    # سيتم حفظ الـ Confusion Matrix بهذا الاسم
    CONFUSION_MATRIX_PNG = PLOTS_DIR / "confusion_matrix.png"

    # ========================================================
    # إنشاء مصفوفة الارتباك الأساسية
    # ========================================================

    # confusion_matrix():
    #
    # دالة من sklearn
    #
    # تقوم بمقارنة:
    #
    # all_labels:
    # التصنيفات الحقيقية
    #
    # all_predictions:
    # توقعات النموذج
    #
    # الناتج:
    # جدول يوضح:
    # - عدد التوقعات الصحيحة
    # - عدد الأخطاء
    cm = confusion_matrix(
        all_labels,
        all_predictions
    )

    # ========================================================
    # إنشاء نافذة تحتوي على رسمين
    # ========================================================

    # الرسم الأول:
    # Confusion Matrix الأصلية
    #
    # الرسم الثاني:
    # Normalized Confusion Matrix
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # ========================================================
    # الرسم الأول — المصفوفة الأساسية
    # ========================================================

    # ConfusionMatrixDisplay:
    #
    # كلاس جاهز لرسم الـ Confusion Matrix
    #
    # confusion_matrix=cm:
    # تمرير بيانات المصفوفة
    #
    # display_labels:
    # أسماء الفئات
    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=target_names
    )

    # ========================================================
    # رسم المصفوفة
    # ========================================================

    # cmap='Blues':
    # استخدام تدرج اللون الأزرق
    #
    # ax=axes[0]:
    # الرسم داخل أول Graph
    #
    # values_format='d':
    # عرض القيم كأرقام صحيحة Integer
    disp.plot(
        cmap='Blues',
        ax=axes[0],
        values_format='d'
    )

    # ========================================================
    # عنوان الرسم الأول
    # ========================================================

    axes[0].set_title(
        "Confusion Matrix",
        fontsize=13,
        fontweight='bold'
    )

    # ========================================================
    # إنشاء النسخة المُعيَّرة Normalized
    # ========================================================

    # cm.astype('float'):
    # تحويل القيم إلى Float
    #
    # cm.sum(axis=1):
    # حساب مجموع كل صف
    #
    # np.newaxis:
    # إعادة تشكيل الأبعاد
    #
    # الهدف:
    # تحويل القيم إلى نسب مئوية
    #
    # مثال:
    #
    # بدل:
    # 95 صورة
    #
    # تصبح:
    # 95%
    cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    # ========================================================
    # إنشاء الرسم الثاني
    # ========================================================

    disp2 = ConfusionMatrixDisplay(
        confusion_matrix=cm_norm,
        display_labels=target_names
    )

    # ========================================================
    # رسم المصفوفة المُعيَّرة
    # ========================================================

    # cmap='Greens':
    # استخدام اللون الأخضر
    #
    # values_format='.2%':
    # عرض القيم كنسب مئوية
    disp2.plot(
        cmap='Greens',
        ax=axes[1],
        values_format='.2%'
    )

    # ========================================================
    # عنوان الرسم الثاني
    # ========================================================

    axes[1].set_title(
        "Normalized Confusion Matrix (%)",
        fontsize=13,
        fontweight='bold'
    )

    # ========================================================
    # العنوان الرئيسي للشكل بالكامل
    # ========================================================

    # accuracy:
    # عرض الدقة الكلية للنظام
    fig.suptitle(
        f"AI Vision System Evaluation — Overall Accuracy: {accuracy:.2f}%",
        fontsize=15,
        fontweight='bold'
    )

    # ========================================================
    # ضبط المسافات بين الرسومات
    # ========================================================

    plt.tight_layout()

    # ========================================================
    # حفظ الصورة
    # ========================================================

    # dpi=200:
    # جودة الصورة
    #
    # bbox_inches='tight':
    # إزالة الفراغات الزائدة
    plt.savefig(
        CONFUSION_MATRIX_PNG,
        dpi=200,
        bbox_inches='tight'
    )

    # ========================================================
    # طباعة رسالة نجاح الحفظ
    # ========================================================

    print(
        f"Confusion matrix saved at: "
        f"{CONFUSION_MATRIX_PNG}"
    )

    # ========================================================
    # عرض الرسومات
    # ========================================================

    plt.show()

    # ========================================================
    # تفسير عناصر الـ Confusion Matrix
    # ========================================================

    # شكل المصفوفة الثنائية:
    #
    #            Predicted
    #            0      1
    #
    # Actual 0  TP     FN
    # Actual 1  FP     TN

    # ========================================================
    # استخراج القيم الأربع الأساسية
    # ========================================================

    # TP:
    # True Positive
    #
    # القطع المعيبة التي اكتشفها النظام بشكل صحيح
    TP = cm[0, 0]

    # FN:
    # False Negative
    #
    # القطع المعيبة التي فشل النظام في اكتشافها
    #
    # أخطر نوع أخطاء في الصناعة
    #
    # لأن القطعة المعيبة ستصل للعميل
    FN = cm[0, 1]

    # FP:
    # False Positive
    #
    # القطع السليمة التي تم رفضها بالخطأ
    FP = cm[1, 0]

    # TN:
    # True Negative
    #
    # القطع السليمة التي تم قبولها بشكل صحيح
    TN = cm[1, 1]

    # ========================================================
    # عرض تفسير النتائج
    # ========================================================

    print(f"\nInspection Results Interpretation:")

    # عدد القطع المعيبة المكتشفة بنجاح
    print(
        f"Correctly Detected Defective Parts (TP): "
        f"{TP} images"
    )

    # عدد القطع السليمة المقبولة بشكل صحيح
    print(
        f"Correctly Accepted Good Parts (TN): "
        f"{TN} images"
    )

    # عدد القطع المعيبة التي لم يكتشفها النظام
    print(
        f"Missed Defective Parts (FN): "
        f"{FN} images ← Most Dangerous Error!"
    )

    # عدد القطع السليمة التي تم رفضها
    print(
        f"Incorrectly Rejected Good Parts (FP): "
        f"{FP} images"
    )

# ============================================================
# معالجة الأخطاء
# ============================================================

except Exception as e:

    # طباعة الخطأ إذا حدثت مشكلة
    print(f"Error: {e}")
````

# استيراد المكتبات وتحميل الموديل

هذه الخلية تدمج المكتبات وتعريف الشبكة العصبية (CNN) وتحميل الأوزان الذكية:

````python
# ============================================================
# 1. استيراد المكتبات الأساسية
# ============================================================
import sys
if hasattr(sys.stdout, "reconfigure"):
    sys.stdout.reconfigure(encoding="utf-8")

import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image
import cv2
import os
import random
import socket
import sys
import msvcrt  # مكتبة لالتقاط ضغطات الكيبورد اللحظية للتحكم بالأنظمة

print("✅ تم تحميل المكتبات بنجاح")

# ============================================================
# 2. بناء معمارية الذكاء الاصطناعي (CNN)
# ============================================================
class DefectDetectionCNN(nn.Module):
    def __init__(self):
        super(DefectDetectionCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(32 * 16 * 16, 128)
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128, 2)

    def forward(self, x):
        x = self.pool1(self.relu1(self.conv1(x)))
        x = self.pool2(self.relu2(self.conv2(x)))
        x = x.view(x.size(0), -1)
        x = self.relu3(self.fc1(x))
        x = self.fc2(x)
        return x

# ============================================================
# 3. تحميل الأوزان المُدربة
# ============================================================
model = DefectDetectionCNN()
# تأكد من مسار ملف الأوزان الخاص بك
model_path = MODEL_PTH if 'MODEL_PTH' in globals() else Path(r"E:\AI\AI project\code\data\robot_vision_brain.pth")

if model_path.exists():
    model.load_state_dict(torch.load(model_path, map_location='cpu'))
else:
    raise FileNotFoundError(f"لم يُعثر على ملف النموذج: {model_path}")
model.eval()
print("✅ الذكاء الاصطناعي جاهز: (AUTO/MANUAL + ARDUINO SYNC)")

# إعداد معالجة الصور
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),
    transforms.Resize((64, 64)),
    transforms.ToTensor()
])
````

# دوال الاتصال وتجهيز البيانات
هذه الخلية تحتوي على دوال إرسال البيانات للأردوينو وتجهيز مسارات الصور:

````python
# ============================================================
# 4. تجهيز الصور لمحاكاة خط الإنتاج
# ============================================================
# عدّل المسارات لتطابق جهازك
folder_ok = r"E:\AI\AI project\code\archive\temp\casting_data\casting_data\train\ok_front"

# مجلد الصور المعيبة (Defective Parts)
folder_defect = r"E:\AI\AI project\code\archive\temp\casting_data\casting_data\train\def_front"

ok_images = [os.path.join(folder_ok, f) for f in os.listdir(folder_ok)]
def_images = [os.path.join(folder_defect, f) for f in os.listdir(folder_defect)]
all_images = ok_images + def_images
random.shuffle(all_images) # خلط الصور لعمل محاكاة واقعية

# ============================================================
# 5. دالة الاتصال اللحظي بـ Arduino Bridge
# ============================================================
def send_to_arduino(label):
    """
    ترسل الأوامر إلى برنامج Bridge الخاص بالأردوينو عبر منفذ 19001.
    يتم إضافة \n في نهاية الأمر ليتعرف الأردوينو على نهاية الرسالة.
    """
    try:
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.connect(('127.0.0.1', 19001)) # البورت المخصص للأردوينو
        
        if label == "DEFECTIVE":
            command = "DEFECT\n"
        elif label == "OK":
            command = "OK\n"
        elif label == "MANUAL":
            command = "MANUAL\n"
        else:
            command = "RESET\n"
            
        s.sendall(command.encode('utf-8'))
        s.close()
        print(f"🔌 Arduino: Sent '{command.strip()}'")
    except Exception:
        pass # التجاوز الصامت في حال لم يكن الأردوينو متصلاً
````

# حلقة التشغيل الرئيسية (The Master Loop)
هذه هي الخلية الأهم والتي تدير النظام بالكامل. يُفضّل عند تشغيلها أن تكون بيئة الكوبيليا والماتلاب و bridge.py في وضع التشغيل:

````python
# ============================================================
# 6. نظام التشغيل المدمج (المايسترو)
# ============================================================
# 1. محاولة الاتصال بـ MATLAB (Master Controller)
try:
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.connect(('127.0.0.1', 65432))
    print("🔗 Successfully Connected to MATLAB TCP Server!")
except Exception as e:
    print("❌ ERROR: MATLAB is not ready. Please run MATLAB first and click 'Connect'.")
    sys.exit()

cv2.namedWindow("AI Conveyor Inspection System")

mode = "AUTO" 
idx = 0       

while idx < len(all_images):
    image_path = all_images[idx]
    
    # ----------------------------------------------------
    # وضع التشغيل الآلي (AUTO Mode)
    # ----------------------------------------------------
    if mode == "AUTO":
        print("\n" + "="*50)
        print("⏳ [AUTO] Waiting for Proximity Sensor... (Press 'M' for Manual, 'Q' to Quit)")
        
        trigger_received = False
        while True:
            # أ. المراقبة اللحظية للوحة المفاتيح (SCADA Commands)
            if msvcrt.kbhit():
                key = msvcrt.getch().decode('utf-8').upper()
                if key == 'Q': # إيقاف الطوارئ
                    try: s.sendall("STOP_SYSTEM\r\n".encode('utf-8')) 
                    except: pass
                    send_to_arduino("RESET")
                    print("\n🛑 Shutting down Python, MATLAB, and Arduino...")
                    sys.exit()
                elif key == 'M': # التحويل للوضع اليدوي
                    mode = "MANUAL"
                    send_to_arduino("MANUAL")
                    print("\n🔄 Switched to MANUAL Mode.")
                    break
            
            # ب. انتظار إشارة الحساس من ماتلاب
            try:
                s.settimeout(0.1)
                msg = s.recv(1024).decode('utf-8')
                if "TRIGGER" in msg:
                    print("🎯 Sensor Trigger Received! Analyzing part...")
                    trigger_received = True
                    break
                elif "QUIT" in msg or "STOP_SYSTEM" in msg:
                    send_to_arduino("RESET")
                    print("\n🛑 MATLAB closed the system. Shutting down...")
                    sys.exit()
            except socket.timeout:
                continue
            except Exception as e:
                print(f"\n❌ Connection Error: {e}")
                sys.exit()
                
        if mode == "MANUAL":
            continue 
            
    # ----------------------------------------------------
    # وضع التشغيل اليدوي (MANUAL Mode)
    # ----------------------------------------------------
    elif mode == "MANUAL":
        print("\n" + "="*50)
        cmd = input("⚙️ [MANUAL] Press [Enter] to process next part, 'A' for AUTO, 'Q' to QUIT: ").strip().upper()
        
        if cmd == 'Q':
            try: s.sendall("STOP_SYSTEM\r\n".encode('utf-8'))
            except: pass
            send_to_arduino("RESET")
            print("🛑 Shutting down Python, MATLAB, and Arduino...")
            sys.exit()
            
        elif cmd == 'A':
            mode = "AUTO"
            send_to_arduino("RESET") 
            print("🔄 Switched to AUTO Mode.")
            continue 

    # ==========================================
    # 7. معالجة الصورة واتخاذ القرار اللحظي
    # ==========================================
    image = cv2.imread(image_path)
    if image is None: 
        idx += 1
        continue
    original = image.copy()

    pil_image = Image.open(image_path)
    input_tensor = transform(pil_image).unsqueeze(0)

    # التنبؤ باستخدام الـ CNN
    with torch.no_grad():
        output = model(input_tensor)
        probabilities = torch.softmax(output, dim=1)
        confidence, predicted = torch.max(probabilities, 1)
        predicted_class = predicted.item()
        confidence_score = confidence.item() * 100

    # تحديد القرار والإحداثيات للروبوت
    if predicted_class == 0:
        label = "DEFECTIVE"
        color = (0, 0, 255)
        target_coords = [250.0, -500.0, 150.0]
        status_code = 0  
    else:
        label = "OK"
        color = (0, 255, 0)
        target_coords = [250.0, 500.0, 150.0]
        status_code = 1  

    h, w, _ = image.shape
    
    # حماية ضد انهيار الرسم (Crash Fix) للقطع الصغيرة
    try:
        cv2.rectangle(original, (5, 5), (max(10, w - 5), max(10, h - 5)), color, 4)
        cv2.putText(original, f"{label} | {confidence_score:.2f}%", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)
    except:
        pass 
    
    cv2.imshow("AI Conveyor Inspection System", original)
    cv2.waitKey(1000)

    print(f"👁️ Vision Result: {label} ({confidence_score:.1f}%)")
    
    # 1. إرسال النتيجة للأردوينو (شاشة الـ LCD)
    send_to_arduino(label)
    
    # 2. إرسال الأوامر للماتلاب لتحريك الروبوت
    try:
        response = f"{target_coords[0]},{target_coords[1]},{target_coords[2]},{status_code}\r\n"
        s.sendall(response.encode('utf-8'))
        print(f"📡 Command Sent to MATLAB! Moving to Y={target_coords[1]}")
    except Exception as e:
        print(f"❌ Failed to send command to MATLAB: {e}")
        break

    idx += 1 

s.close()
cv2.destroyAllWindows()
````

````python
# ============================================================
# الدوال المساعدة للتنسيق
# ============================================================
def autosize(ws, min_w=10, max_w=35):
    """ضبط عرض كل عمود تلقائيًا بناءً على محتواه"""
    for col in ws.columns:
        letter = get_column_letter(col[0].column)
        max_len = max((len(str(c.value)) for c in col if c.value), default=0)
        ws.column_dimensions[letter].width = min(max(max_len + 2, min_w), max_w)


def style_range(ws, header_row=1, start_row=2):
    """تطبيق التنسيق على الرأس والبيانات"""
    for row in ws.iter_rows():
        for cell in row:
            cell.border = border
            cell.alignment = center if cell.column != 1 else left_align
    for cell in ws[header_row]:
        cell.fill = header_fill
        cell.font = header_font
        cell.alignment = center
    autosize(ws)
    ws.row_dimensions[header_row].height = 22
    ws.freeze_panes = f'A{start_row}'  # تجميد صف الرأس عند التمرير


def add_title_band(ws, title):
    """إضافة شريط العنوان الداكن في الأعلى"""
    ws.insert_rows(1)
    ws.merge_cells(start_row=1, start_column=1,
                   end_row=1,   end_column=ws.max_column)
    c = ws.cell(1, 1)
    c.value    = title
    c.fill     = band_fill
    c.font     = Font(color='FFFFFF', bold=True, size=13)
    c.alignment = center
    ws.row_dimensions[1].height = 24


def add_table(ws, name, start_row=2):
    """إضافة جدول Excel قابل للتصفية والفرز"""
    ref = f"A{start_row}:{get_column_letter(ws.max_column)}{ws.max_row}"
    if ws.max_row >= start_row + 1:
        tab = Table(displayName=name, ref=ref)
        tab.tableStyleInfo = TableStyleInfo(
            name='TableStyleMedium2',
            showRowStripes=True,
            showColumnStripes=False
        )
        ws.add_table(tab)

print("✅ تم تعريف دوال التنسيق")
````

````python
# ============================================================
# إنشاء Excel لملف المقارنة (Benchmark)
# ============================================================
def save_pretty_excel_from_csv(csv_path, out_path, sheet_title, table_name,
                                sort_col=None, highlight_best=None):
    if not Path(csv_path).is_file():
        print(f"⚠️ الملف غير موجود: {csv_path}")
        return None

    df = pd.read_csv(csv_path)
    if df.empty:
        print(f"⚠️ الملف فارغ: {csv_path}")
        return None

    if sort_col and sort_col in df.columns:
        df = df.sort_values(sort_col).reset_index(drop=True)

    # إذا الملف موجود — احذفه أولاً
    if Path(out_path).exists():
        os.remove(out_path)

    df.to_excel(out_path, index=False, sheet_name=sheet_title)
    wb = load_workbook(out_path)
    ws = wb[sheet_title]

    add_title_band(ws, sheet_title)
    style_range(ws, header_row=2, start_row=3)
    add_table(ws, table_name)

    # تمييز أفضل صف
    if highlight_best and highlight_best in [ws.cell(2, c).value for c in range(1, ws.max_column+1)]:
        headers = {ws.cell(2, c).value: c for c in range(1, ws.max_column+1)}
        col_idx = headers[highlight_best]
        vals = [(r, ws.cell(r, col_idx).value) for r in range(3, ws.max_row+1)
                if isinstance(ws.cell(r, col_idx).value, (int, float))]
        if vals:
            best_row = min(vals, key=lambda x: x[1])[0]
            for c in range(1, ws.max_column+1):
                ws.cell(best_row, c).fill = best_fill
                ws.cell(best_row, c).font = best_font

    wb.save(out_path)
    wb.close()
    return out_path


# --- تنفيذ الإنشاء ---
ts = datetime.now().strftime('%Y%m%d_%H%M%S')

bench_out = EXCEL_DIR / f"matlab_benchmark_results_{ts}.xlsx"
result = save_pretty_excel_from_csv(
    BENCH_CSV, bench_out,
    'Benchmark Data', 'BenchmarkTable',
    sort_col='cost', highlight_best='cost'
)
if result:
    print(f"✅ تم إنشاء: {result}")
else:
    print("⚠️ لم يُنشأ ملف Benchmark (تحقق من وجود CSV)")

runtime_out = EXCEL_DIR / f"robot_runtime_log_{ts}.xlsx"
result2 = save_pretty_excel_from_csv(
    RUNTIME_CSV, runtime_out,
    'Runtime Log', 'RuntimeLogTable',
    sort_col='timestamp' if RUNTIME_CSV.exists() else None
)
if result2:
    print(f"✅ تم إنشاء: {result2}")
else:
    print("⚠️ لم يُنشأ ملف Runtime")
````
